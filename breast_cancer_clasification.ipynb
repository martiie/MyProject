```python
import numpy as np
import pandas as pd
from pycaret.classification import *
```

```python
dataset = pd.read_csv("/content/breast-cancer.csv")
dataset
```

# Data Preparation

```python
dataset = dataset.dropna()
dataset
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_style = OrdinalEncoder(categories=[['M','B']]) #
results = ordinal_style.fit_transform(dataset[["diagnosis"]])
encoded_df = pd.DataFrame(results, columns=['class'])
dataset = dataset.join(encoded_df)
dataset.head()
```

# Model

```python
classifier = setup(data = dataset, target = 'class',ignore_features = ['id',	'diagnosis'] , session_id=42)
```

```python
compare_models()
```

```python
dt = create_model('rf')
```

```python
X = dataset.drop(['id',	'diagnosis','class'], axis=1)
y = dataset['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
KNN = KNeighborsClassifier()
KNN.fit(X_train,y_train)
y_pred = KNN.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
GaussianNB = GaussianNB()
GaussianNB.fit(X_train,y_train)
y_pred = GaussianNB.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
Decisiontree = DecisionTreeClassifier(criterion = 'entropy')
Decisiontree.fit(X_train,y_train)
y_pred = Decisiontree.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators = 1000)
RF.fit(X_train,y_train)
y_pred = RF.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from numpy import mean

cv = KFold(n_splits = 10,random_state=1,shuffle=True)
scores = cross_val_score(RF,X_train,y_train,scoring = 'accuracy',cv=cv,n_jobs=1)
print(mean(scores))
```
